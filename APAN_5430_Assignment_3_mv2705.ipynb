# 4. Use NTLK and/or Spacy  tokenization features to: List and count n-grams for any given input n


```python
import re
from nltk.util import ngrams
import collections

def ngram_finder(s,n):
    s = s.lower()
    s = re.sub(r'[^a-zA-Z0-9\s]', ' ', s)
    tokens = [token for token in s.split(" ") if token != ""]
    output = list(ngrams(tokens, n))
    #return output
    #print(output)
    result = collections.Counter(output)
    return result
```

```python
ngram_counted = (ngram_finder(url_text,2))
```

```python
ngram_counted.most_common()
```

# 4. Use NTLK and/or Spacy  tokenization features to: Lemmatize and deduplicate unigrams into a vocabulary of terms.

```python
from nltk.stem import WordNetLemmatizer

text_tokenize = nltk.word_tokenize(url_text)
stop = stopwords.words('english') + list(string.punctuation)
vocab = [i for i in text_tokenize if i not in stop]
lemmatizer = WordNetLemmatizer()
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in vocab])
print(lemmatized_output)
```

```python
lemmatized_output1 = [lemmatizer.lemmatize(w) for w in vocab]
print(set(lemmatized_output1))
```
