# Gathering Article from the NYT API

### The Below code tests a users' access to the requests library to access NYT Meta-data for a querieble search such as "Markets". 

### User can input their given search string as parameter 'q', along with a 'begin date' and a sort function. 

### The code below access the api searching for 'Markets', and returns article headlines that have been returned from the search. 

```python
import requests
queries = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q='Markets'&begin_date=20190923&sort=newest&api-key=3GAhTXvfqXPEQIO3UEZLVTxr7TF1AsD0"
articles = requests.get(queries).json()
for i in articles['response']['docs']: #print headlines (check relevancy)
    print(i['headline']['main'])
```

### If the search returns results, we can go ahead and iterate through the different pages of the api search, to access more articles from the search. 

### This is done in the function defined below. Please note, that '.format' function is called to iterate through each page of the API.  Sleep is called to ensure that we are scraping within large time intervals, while the 'if' clause pulls if the search is successful. 

### We will choose an arbitrary date to start the search, and further refine this search parameter so that we pull from the dates sequentially by running the search again using the last dated article from the api pull

### Moreover, the api only returns 200 pages of searchable articles. Thus, we must search in increments of 200.


```python
from random import randint
from time import sleep
from tqdm import tqdm_notebook as tqdm
import pickle 



url = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q='Markets'&page={}&begin_date=20100103&sort=newest&api-key=3GAhTXvfqXPEQIO3UEZLVTxr7TF1AsD0"
def nyt_get(url):
    #begin_date=20100103 is arbitrary for our first pull. 
        #we will further change this search paramter later on. 
    nyt_list = []

        
    for i in tqdm(range(0,200)):

        #url = url.format(i)
        r = requests.get(url.format(i))
        sleep(6)
        print(url.format(i))
        print(r.status_code)
        if r.status_code == 200:
            
            nyt_list.append(r.json())
            with open('nyt_list_nov_05.pkl', 'wb') as f:
                pickle.dump(nyt_list, f)
            
        else:
            print(r.status_code)
            print('Can No longer Pull')

    return nyt_list
```

```python
nyt_list_1 = nyt_get(url)
```
