# Sort the dictionary by the "rank" field in ascending order:
## Copying values into an array or list, and sorting, or
### Using Python sorted function on the dictionary

```python
sorted_df_dict = sorted(df_dict, key = lambda i: i['rank'],reverse=False)
```

# MJV COMMENTS AS PER JOHN's INSTRUCTIONS

#### The lambda function is an anonymous function in python, rather than a normal function that uses the 'def' preface

#### This function can have any number of arguments but only one expression, which is evaluated and returned. Moreover, lambda functions are syntactically restricted to a single expression.

#### sorted() built-in function that builds a new sorted list from an iterable. 

#### key parameter to specify a function to be called on each list element prior to making comparisons.
